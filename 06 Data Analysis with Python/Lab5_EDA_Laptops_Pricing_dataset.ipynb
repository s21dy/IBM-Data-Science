# GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.

```python
# Group by 
group_df = df[[ "GPU", "CPU_core", "Price"]].groupby(["GPU", "CPU_core"]).mean()
group_df.reset_index(inplace = True)

# Pivot 
pivot_tb = group_df.pivot(index = "GPU", columns = ["CPU_core"])
```

```python
pivot_tb
```

```python
# Creating pcolor plot 
fig, ax = plt.subplots()
im = ax.pcolor(pivot_tb, cmap='RdBu')

row_labels = pivot_tb.columns.levels[1]
col_labels = pivot_tb.index

#move ticks and labels to the center
ax.set_xticks(np.arange(pivot_tb.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(pivot_tb.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

plt.show()

#rotate label if too long
fig.colorbar(im)
plt.show()
```

# Pearson Correlation and p-values

```python
pearsonr_corr, p_val = stats.pearsonr(df["GPU"], df['Price'])
print("Pearson Correlation: ",  pearsonr_corr)
print("P-Value: ",  p_val)
```

```python
pearsonr_corr, p_val = stats.pearsonr(df["CPU_core"], df['Price'])
print("Pearson Correlation: ",  pearsonr_corr)
print("P-Value: ",  p_val)
```
