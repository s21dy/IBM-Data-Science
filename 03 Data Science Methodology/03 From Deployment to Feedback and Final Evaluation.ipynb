# Deployment

## Overview
- Objective: To make a data science model relevant and useful by familiarizing stakeholders with the tool produced.
- Key Stakeholders: Solution owner, marketing, application developers, and IT administration.

## Deployment Process
1. Evaluation and Confidence Building:
    - The model is evaluated by the data scientist.
    - If the model is deemed reliable, it is deployed.
    - Initial deployment might be limited to a small group or test environment to build confidence.
    
## Case Study: Reducing Readmission Risk
- Preparation:
    - Assimilate knowledge for the business group designing the intervention program.
    - Translate model results for clinical staff to identify high-risk patients and design interventions.
- Requirements:
    - Application for automated, near real-time risk assessments of congestive heart failure.
    - User-friendly, browser-based application on tablets for clinical staff.  
- Data Handling:
    - Patient data generated during the hospital stay.
    - Data prepared in a format needed by the model.
    - Each patient scored near discharge for up-to-date risk assessment.
- Deployment Activities:
    - Develop and deliver training for clinical staff.
    - Create processes for tracking and monitoring patients in collaboration with IT developers and database administrators.
    
- Example Solution: Cognos Application for Juvenile Diabetes
    - Risk Model: Decision tree classification for hospitalization risk.
        - Interactive Map:
            - Overview of hospitalization risk nationwide.
            - Interactive analysis of predicted risk by various patient conditions and characteristics.
        - Summary Reports:
            - Summary report of risk by patient population within a model node.
            - Detailed summary for individual patients, including predicted risk and clinical history for doctors.
            

# Feedback

## Overview
- Objective: To refine the model and assess its performance and impact through user feedback.
- Importance: Incorporating feedback and making adjustments ensures the model remains valuable as long as the solution is needed.

## Feedback Process
1. Cyclical Methodology:
    - Each step in the Data Science Methodology sets the stage for the next.
    - Ensures continuous refinement at each stage.

2. John Rollins' Perspective:
    - The feedback process is based on the notion that gaining more knowledge leads to the desire for even more understanding.
    
## Case Study: Congestive Heart Failure Risk Population
- Feedback Stage Plan:
1. Review Process:
    - Define and implement a review process.
    - Clnical management executives are responsible for measuring the results of the "flying to risk" model.
2. Tracking Outcomes:
    - Track congestive heart failure patients receiving intervention.
    - Record their re-admission outcomes.
    
3. Measure Effectiveness:
- Measure the intervention's effectiveness in reducing re-admissions.
- Ethical approach: Compare readmission rates before and after model implementation, rather than using controlled and treatment groups.

## Refinement and Impact Assessment
- First Year Review:
    - After one year of implementation, review the impact on re-admission rates.
    - Refine the model based on compiled data and knowledge gained.

- Further Refinements:
    - Incorporate additional information, such as participation in the intervention program and detailed pharmaceutical data.
    - Address the possibility of adding pharmaceutical data if deemed valuable after feedback.
    - Allow for other refinements that may arise during the feedback stage.

- Review and Refine Actions
    - Review and refine intervention actions and processes based on deployment and feedback experience.

## Redeployment
- Continuous Feedback:
    - Redeploy the refined model and intervention actions.
    - Continue the feedback process throughout the life of the intervention program.

# Storytelling

## Importance of Storytelling
- Essential Skill: Storytelling with data is critical for data analysts.
- Human Understanding: Humans naturally understand the world through stories.

## Convincing with Data
- Clear, Concise, Compelling: To convince others, a clear, concise, and compelling story is necessary.
- Self-Understanding: Developing a story helps data analysts understand the dataset better.

## Balancing Simplicity and Complexity
- Clear and Coherent: Finding the balance between a simple story and the complexities within the data is challenging but crucial.
- Communication: Effective communication of data insights is vital, regardless of the audience level.

## Communicating Value
- Visuals and Stories: Communicating data effectively often involves visuals and storytelling.
- Audience Understanding: Ensuring the audience understands how the information can be useful is key.

## Storytelling as a Skill Set
- Essential Skill: Storytelling is like the last mile in delivery.
- Technical vs. Storytelling Skills: Technical skills can be trained, but the ability to extract and communicate value is rarer.

## Long-term Career
- Critical Skill: Knowing how to tell a compelling story with data is crucial for a long-term career in data analytics.

## Conveying the Message
- Resonance: Numbers alone are not enough; a story is needed to make the message resonate.
- Emotional Connection: An emotional connection through storytelling drives home the message and prompts action.

## Stanford Study
- Study Results: A study showed that stories in presentations are more memorable than just KPIs and statistics.
- Emotional Impact: Facts and figures within a story create an emotional connection, making the data more impactful.

# Glossary

| Term | Definition |
|:-:|:---|
| Binary classification model | A model that classifies data into two categories, such as yes/no or stop/go outcomes. |
| Data compilation | The process of gathering and organizing data required for modeling. |
| Data modeling | The stage in the data science methodology where data scientists develop models, either descriptive or predictive, to answer specific questions. |
| Descriptive model | A type of model that examines relationships between variables and makes inferences based on observed patterns. |
| Diagnostic measure based tuning | The process of fine-tuning the model by adjusting parameters based on diagnostic measures and performance indicators. |
| Diagnostic measures | The evaluation of a model's performance of a model to ensure that the model functions as intended. |
| Discrimination criterion | A measure used to evaluate the performance of the model in classifying different outcomes. |
| False-positive rate | The rate at which the model incorrectly identifies negative outcomes as positive. |
| Histogram | A graphical representation of the distribution of a dataset, where the data is divided into intervals or bins, and the height of each bar represents the frequency or count of data points falling within that interval. |
| Maximum separation | The point where the ROC curve provides the best discrimination between true-positive and false-positive rates, indicating the most effective model. |
| Model evaluation | The process of assessing the quality and relevance of the model before deployment. |
| Optimal model | The model that provides the maximum separation between the ROC curve and the baseline, indicating higher accuracy and effectiveness. |
| Receiver Operating Characteristic (ROC) | Originally developed for military radar, the military used this statistical curve to assess the performance of binary classification models. |
| Relative misclassification cost | This measurement is a parameter in model building used to tune the trade-off between true-positive and false-positive rates. |
| ROC curve (Receiver Operating Characteristic curve) | A diagnostic tool used to determine the optimal classification model's performance. |
| Separation | Separation is the degree of discrimination achieved by the model in correctly classifying outcomes. |
| Statistical significance testing | Evaluation technique to verify that data is appropriately handled and interpreted within the model. |
| True-positive rate | The rate at which the model correctly identifies positive outcomes. |
